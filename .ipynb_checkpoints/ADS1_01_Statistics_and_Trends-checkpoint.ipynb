```python
#importing models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing datas
df=pd.read_csv("Datas/Data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(["Country Code", "Series Code", "2021 [YR2021]", "2022 [YR2022]"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.rename(columns={"2008 [YR2008]":"2008","2009 [YR2009]":"2009", "2010 [YR2010]":"2010", "2011 [YR2011]":"2011", "2012 [YR2012]":"2012", "2013 [YR2013]":"2013", "2014 [YR2014]":"2014", "2015 [YR2015]":"2015", "2016 [YR2016]":"2016", "2017 [YR2017]":"2017", "2018 [YR2018]":"2018", "2019 [YR2019]":"2019", "2020 [YR2020]":"2020"}, inplace=True)
```

```python
df.head()
```

```python
#Plottong Bars graph based on Agricultural nitrous oxide emissions (thousand... 

selected_series = "Agricultural nitrous oxide emissions (thousand metric tons of CO2 equivalent)"
filtered_df = df[df["Series Name"] == selected_series]


# Extract relevant columns for plotting
countries = filtered_df["Country Name"]
years = df.columns[2:]

# Plotting the Bar Plot
plt.figure(figsize=(12, 6))
for i in range(len(filtered_df)):
    plt.bar(years, filtered_df.iloc[i, 2:], label=countries.iloc[i])

plt.xlabel("Year")
plt.ylabel("Value")
plt.title(selected_series)
plt.legend()
plt.show()


```

```python
#Plottong Histogram graph based on Agricultural land (% of land area) 

selected_series = "Agricultural land (% of land area)"
filtered_df = df[df["Series Name"] == selected_series]


# Extract relevant columns for plotting
countries = filtered_df["Country Name"]
years = df.columns[2:]

# Plotting the histograms
plt.figure(figsize=(12, 6))
for i in range(len(filtered_df)):
    plt.bar(years, filtered_df.iloc[i, 2:], label=countries.iloc[i])

plt.xlabel("Year")
plt.ylabel("Value")
plt.title(selected_series)
plt.legend()
plt.show()
```

```python
#Plottong Histogram graph based on CO2 emissions (kt) 

selected_series = "CO2 emissions (kt)"
filtered_df = df[df["Series Name"] == selected_series]


# Extract relevant columns for plotting
countries = filtered_df["Country Name"]
years = df.columns[2:]

# Plotting the histograms
plt.figure(figsize=(12, 6))
for i in range(len(filtered_df)):
    plt.bar(years, filtered_df.iloc[i, 2:], label=countries.iloc[i])

plt.xlabel("Year")
plt.ylabel("Value")
plt.title(selected_series)
plt.legend()
plt.show()
```

```python
#Plottong Histogram graph based on Population growth (annual %) 

selected_series = "Population growth (annual %)"
filtered_df = df[df["Series Name"] == selected_series]


# Extract relevant columns for plotting
countries = filtered_df["Country Name"]
years = df.columns[2:]

# Plotting the histograms
plt.figure(figsize=(12, 6))
for i in range(len(filtered_df)):
    plt.bar(years, filtered_df.iloc[i, 2:], label=countries.iloc[i])

plt.xlabel("Year")
plt.ylabel("Value")
plt.title(selected_series)
plt.legend()
plt.show()
```

```python
# Extract relevant columns for plotting
countries = df["Country Name"].unique()
years = df.columns[2:]

# Plotting the line plot
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = df[df["Country Name"] == country]
    plt.plot(years, country_data.iloc[0, 2:], label=country)

plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Line Plot for the Dataset")
plt.legend()
plt.show()
```

```python

# Exclude non-numeric columns for plotting
numeric_data = df.iloc[:, 2:].astype(float)

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data, cmap="YlGnBu", xticklabels=years, yticklabels=countries, annot=True, fmt=".2f", linewidths=.5)
plt.xlabel("Year")
plt.ylabel("Country")
plt.title("Heatmap for the Dataset")
plt.show()
```

```python

# Pivot the DataFrame to create a heatmap
heatmap_data = df.pivot(index='Country Name', columns='Series Name', values=df.columns[2:])

# Convert the values to numeric (in case they are not already)
heatmap_data = heatmap_data.apply(pd.to_numeric, errors='coerce')

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=.5)
plt.xlabel("Series Name")
plt.ylabel("Country Name")
plt.title("Heatmap for the Dataset")
plt.show()
```
